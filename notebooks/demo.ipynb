```
from pun_detector.agent import PunAgent
from pun_detector.aoa import AoAIndex

# Optional: load AoA from your local file (xlsx/csv).
# aoa = AoAIndex.from_excel('data/AoA_ratings_Kuperman_et_al_BRM_with_PoS.xlsx')
# agent = PunAgent(aoa=aoa)

agent = PunAgent(aoa=None)
```

```
text = "Why did the math book look sad? Because it had too many problems."
age = 12
print(agent.analyze(text, age=age, reveal_scores=False))
```

```
# If you explicitly want to see similarity scores:
print(agent.analyze(text, age=age, reveal_scores=True))
```
